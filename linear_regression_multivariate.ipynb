```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Exercise/hiring.csv")
data
```

```python
data.experience.fillna("one" , inplace = True)
data["test_score(out of 10)"].fillna(data["test_score(out of 10)"].min() , inplace = True)
data
```

```python
def numirc(x):
    if x == "one":
        return 1
    elif x == "two":
        return 2
    
    elif x == "three":
        return 3
    
    elif x == "four":
        return 4
    
    elif x == "five":
        return 5
    
    elif x == "six":
        return 6
    
    elif x == "seven":
        return 7
    
    elif x == "eight":
        return 8
    
    elif x == "nine":
        return 9
    
    elif x == "ten":
        return 10
    
    elif x == "eleven":
        return 11
    
    elif x == "twelve":
        return 12
```

```python
data.experience = data.experience.apply(numirc)
data
```

```python
x = data.drop("salary($)"  , axis = 1)
y = data["salary($)"]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[1,8,9]])
```

```python
modelpredictions = model.predict(x)
modelpredictions
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error

mean_absolute_error(y,modelpredictions)
```

```python
mean_squared_error(y, modelpredictions)
```
